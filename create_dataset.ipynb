```python
import os
import pandas as pd
import dataset as dataset
```

The data frame is created by walking through the dataset directory and using the get_stream_as_df() function to extract the features and the data

```python
dataset_df = pd.D
ataFrame()
dir_basepath = "./dataset"
ignore = ["groundtruth", "faces", ".ipynb_checkpoints", "dataset", "data"]
for subdir, dirs, files in os.walk(dir_basepath):
    if files:
        if os.path.basename(subdir) in ignore:
            continue
        name= os.path.basename(subdir)
        df = dataset.get_stream_as_df(name)
        df["stream"] = name
        df["sequance"] = name[:-3]
        
        dataset_df = pd.concat([dataset_df, df], ignore_index=True)
        filepaths = [os.path.join(subdir, f) for f in files][:10]
```
